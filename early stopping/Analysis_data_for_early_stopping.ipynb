```python
#HEK-NIH
import matplotlib.pyplot as plt
import numpy as np

# Updated Data
models = ['scVAG+', 'scVAG', 'scGAC', 'SCEA', 'scGNN', 'DESC', 'Seurat', 'scASGC', 'SC3', 'DRJCC', 'JSRC', 'scLDS2', 'sLMIC', 'NIC']
runtime = np.array([4680, 21256, 20099, 21530, 6321, 4312, 1342, 9100, 5501, 6312, 28800, 9600, 24520, 5284])
ari = np.array([0.811, 0.81, 0.792, 0.793, 0.309, 0.291, 0.264, 0.794, 0.718, 0.7485, 0.801, 0.665, 0.8168, 0.7993])
nmi = np.array([0.888, 0.891, 0.732, 0.708, 0.524, 0.525, 0.506, 0.859, 0.646, 0.6545, 0.739, 0.644, 0.609, 0.71])

# Create color map
colors = plt.cm.rainbow(np.linspace(0, 1, len(models)))

# Function to plot and annotate with flipped axes and proper scaling
def plot_and_annotate(ax, x, y, colors, models, xlabel, ylabel, title):
    for i, model in enumerate(models):
        ax.scatter(x[i], y[i], color=colors[i], s=150, label=model)  # Increased point size for clarity
        ax.annotate(model, (x[i], y[i]), xytext=(7, 7), textcoords='offset points', fontsize=10, fontweight='bold')

    ax.set_yscale('log')
    ax.set_xlabel(xlabel, fontsize=14)
    ax.set_ylabel(ylabel, fontsize=14)
    ax.set_title(title, fontsize=16)
    ax.grid(True, which="both", ls="-", alpha=0.2)
    ax.tick_params(axis='both', which='major', labelsize=12)

    # Increase the limits of the x and y axes to avoid overlapping
    ax.set_xlim(min(x) - 0.05, max(x) + 0.05)  # Adding padding to x-axis
    ax.set_ylim(min(y) * 0.9, max(y) * 1.1)    # Adjust y-axis log scale padding

    # Draw a vertical line at 0.8
    ax.axvline(x=0.8, color='red', linestyle='--', lw=2)

# Plot for ARI vs Runtime
fig_ari, ax1 = plt.subplots(figsize=(14, 10))  # Increased figure size for better scaling
plot_and_annotate(ax1, ari, runtime, colors, models, 'ARI', 'Runtime (s) - Log Scale', 'ARI vs Runtime')

# Save ARI plot as an SVG file
fig_ari.savefig('ari_vs_HEK-NIH.svg', format='svg', bbox_inches='tight')

# Show ARI plot
plt.show()

# Plot for NMI vs Runtime
fig_nmi, ax2 = plt.subplots(figsize=(14, 10))  # Increased figure size for better scaling
plot_and_annotate(ax2, nmi, runtime, colors, models, 'NMI', 'Runtime (s) - Log Scale', 'NMI vs Runtime')

# Save NMI plot as an SVG file
fig_nmi.savefig('nmi_vs_HEK-NIH.svg', format='svg', bbox_inches='tight')

# Show NMI plot
plt.show()
```
